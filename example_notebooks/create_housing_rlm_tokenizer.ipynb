```python
from tokenizers import Tokenizer
from krk_ml_utils.preprocessing import generate_p10_special_tokens

# 1. Load the pretrained BERT case-sensitive tokenizer
tokenizer = Tokenizer.from_pretrained("roberta-base")

# 2. Define and add your special tokens

special_tokens = generate_p10_special_tokens()

print(f"Number of tokens in the tokenizer before adding special tokens: {tokenizer.get_vocab_size()}")
print(f"Special tokens to be added: {special_tokens}")
print(f"Number of special tokens: {len(special_tokens)}")

tokenizer.add_tokens(special_tokens)

# 3. Save the tokenizer to a file
tokenizer.save("rlm_tokenizer.json")

for token in special_tokens:
    print(f"Token: {token}, ID: {tokenizer.token_to_id(token)}")
    
print(f"Number of tokens in the tokenizer after adding special tokens: {tokenizer.get_vocab_size()}")
```
